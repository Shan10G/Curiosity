```python
# Task : 
# group by Location, find sum of TS and sum of OC

# Step 1 :
temp = stores.groupby("Location")

# Step 2 :
temp = temp[["TotalSales","OperatingCost"]]

# Step 3 : sum of both
Result1 = temp.sum().add_prefix("Sum_")

display(Result1)

# Step 4 : Resetting index
Result2 = Result1.reset_index()
print("--------------------------------------------------------")
display(Result2)
```

```python
# Task : 
# Group by Location, find sum of OC, mean of TS, count Location

# Step 1 : 
temp = stores.groupby("Location")

# Step 2
# 2.1 for Location count
a = temp[["Location"]].count().add_prefix("Count_")
#display(a)
# 2.2 for TotalSales mean
b = temp[["TotalSales"]].mean().add_prefix("Mean_")
#display(b)
# 2.3 for Operating Cost sum
c = temp[["OperatingCost"]].sum().add_prefix("Sum_")
#display(c)

Result1 = pd.concat([a,b,c],axis=1)
display(Result1)

Result2 = Result1.reset_index()
display(Result2)
```
